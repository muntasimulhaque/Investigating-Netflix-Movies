```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

```python
# Displays the first 5 rows of the DataFrame by default
netflix_df.head()
```

```python
# Provides a summary of the DataFrame
netflix_df.info()
```

```python
# Filter for movies for the 1990s
movies_1990s = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]
```

```python
# checking if it worked
movies_1990s.head()
```

```python
# finding the most frquest duration of the 1990s movies
duration_count = movies_1990s["duration"].value_counts()
```

```python
# checking whether it worked
duration_count.head()
```

```python
# The most frequent movie duration in the 1990s
duration = duration_count.idxmax()
```

```python
# checking if it worked
print(duration)
```

```python
# cross-checking by plotting a histogram
movies_1990s['duration'].plot(kind='hist', bins=20)
plt.show()
```

```python
# Filter for short action movies releasd in 1990s
short_action_movies_1990s = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['genre'] == 'Action') & (netflix_df['duration'] < 90) & (netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]
```

```python
#checking if it worked
short_action_movies_1990s.head()
```

```python
# counting the number of short action movies in 1990s
short_movie_count = len(short_action_movies_1990s)
```

```python
#printing the number
print(short_movie_count)
```
